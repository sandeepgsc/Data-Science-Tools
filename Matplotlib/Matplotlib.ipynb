## Matplotlib 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,25,3)
y= [9,6,7,4,65,3,3,10]
```

### scatter_plot

```python
plt.xlabel('X_axis')
plt.ylabel('y_axis')
plt.title('Scatter_plot')
plt.scatter(x,y,color='red',marker='+')
plt.show()
```

### Plotting_technique

```python
plt.xlabel('X_axis')
plt.ylabel('y_axis')
plt.title('plotting_techniques')
plt.plot(x,y,color='Blue',marker='*')
plt.show()
```

#### Legend

```python
days=[1,2,3,4,5,6,7]
a=[50,51,52,48,47,49,46]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,a)
plt.plot(days,b)
plt.plot(days,c)
```

```python
plt.plot(days,a,label = 'max',color='red')
plt.plot(days,b,label = 'min',color='blue')
plt.plot(days,c ,label = 'Avg')
plt.legend(loc='best')
```

```python
plt.plot(days,a,label = 'max',color='red')
plt.plot(days,b,label = 'min',color='blue')
plt.plot(days,c ,label = 'Avg')
plt.legend(loc=1,shadow=True)
```

```python
plt.plot(days,a,label = 'max',color='red')
plt.plot(days,b,label = 'min',color='blue')
plt.plot(days,c ,label = 'Avg')
plt.legend(loc=1,shadow=True,fontsize='large')
```

```python
plt.plot(days,a,label = 'max',color='red')
plt.plot(days,b,label = 'min',color='blue')
plt.plot(days,c ,label = 'Avg')
plt.legend(loc=3,shadow=True,fontsize='xx-large')
```

### Bar_plot

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
a = np.arange(len(company))
a
```

```python
plt.bar(a,revenue,label='Revenue')
plt.title('Bar_plot')
plt.xticks(a,company)
plt.ylabel('Revenue')
plt.legend(loc=1)
plt.show()
```

```python
plt.barh(a,revenue,label='Revenue')
plt.title('Bar_plot')
plt.yticks(a,company)
plt.xlabel('Revenue')
plt.legend(loc=1)
plt.show()
```

## Histogram

```python
a = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(a,bins=40) 
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(a, bins=[80,100,125,150], rwidth=0.9, color='r')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
a = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
b = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([a,b], bins=[80,100,125,150], rwidth=None, color=['green','orange'],label=['men','women'])
plt.legend(loc=1,fontsize='xx-large')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
a = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
b = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([a,b], bins=[80,100,125,150], rwidth=None, color=['green','orange'],label=['men','women'],orientation='horizontal')
plt.legend(loc=1,fontsize='xx-large')
```

## Pie_chart

```python
x = [40,30,20,10,1]
y = ['New_york','Washington','Delas','India','south_Africa']
plt.pie(x,labels=y)
```

```python
x = [40,30,20,10,1]
y = ['New_york','Washington','Delas','India','south_Africa']
plt.pie(x,labels=y,shadow = True)
```

```python
x = [40,30,20,10,1]
y = ['New_york','Washington','Delas','India','south_Africa']
plt.pie(x,labels=y,shadow=True,autopct='%1.1f%%')
```
